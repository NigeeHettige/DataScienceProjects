```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('Housing.csv')
```

```python
df.head(3)
```

```python
new_df = pd.DataFrame()
new_df[['Area','Price']] = df[['area','price']]
```

```python
new_df.head(3)
```

```python
plt.xlabel("Area(sqr ft)")
plt.ylabel("Price(US$)")
plt.scatter(new_df.Area,new_df.Price,color='red',marker='+')
plt.show()
```

```python
reg = LinearRegression()
```

```python
reg.fit(new_df[['Area']],new_df.Price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
